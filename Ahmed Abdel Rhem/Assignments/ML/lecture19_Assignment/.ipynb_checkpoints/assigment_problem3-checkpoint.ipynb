```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
df.info()
```

```python
dict = {'F' : 1, 'M' : 2}
df['Gender_type'] = df['Gender'].map(dict)
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0],inplace=True)
df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0],inplace=True)
df.isnull().sum()
```

```python
dict = {'A' : 1, 'B' : 2,'C':3}
df['City_Category_type'] = df['City_Category'].map(dict)
df.head()
```

```python
df=pd.get_dummies(df,columns=['Age','Stay_In_Current_City_Years'],drop_first=True)
df.head()
```

```python
df.drop(['Gender','City_Category','User_ID','Product_ID'],axis=1,inplace=True)
df.head()
```

```python
x=df[df.columns.difference(['Purchase'])]
x
```

```python
y=df['Purchase']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from xgboost import XGBRFRegressor
```

```python
models={
    'LR':LinearRegression(),
    'KNN':KNeighborsRegressor(),
    'SVR':SVR(),
    'DT':DecisionTreeRegressor(),
    'XGBR':XGBRFRegressor(),
    'RF':RandomForestRegressor()}
```

```python
for name,model in models.items():
    print(f'using:{name}')
    model.fit(x_train,y_train)
    print(f'Train Accuracy:{model.score(x_train,y_train)}')
    print(f'Test Accuracy:{model.score(x_test,y_test)}')
    print('-'*30,'\n')
```

```python
model=RandomForestRegressor()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
import joblib
joblib.dump(model,'model3.h5')
joblib.dump(scaler,'scaler3.h5')
```

```python
m=joblib.load('model3.h5')
s=joblib.load('scaler3.h5')
```

```python
data=s.transform([[1,0,0,0,0,0,2,2,1,15,8,14,14,1,0,0,0]])
m.predict(data)
```
