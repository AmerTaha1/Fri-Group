```python
df=pd.get_dummies(df,columns=['purpose'],drop_first=True)
df.head()
```

```python
sns.boxplot (x='installment',data=df)
sns.stripplot (x='installment',data=df)
```

```python
sns.boxplot(x='int.rate',data=df)
sns.stripplot(x='int.rate',data=df)
```

```python
cols = ['installment', 'int.rate']

from datasist.structdata import detect_outliers

for col in cols:
    indices=detect_outliers(df,0,[col])
    df.loc[indices,col]=df[col].median()
```

```python
sns.boxplot(x='int.rate',data=df)
sns.stripplot(x='int.rate',data=df)
```

```python
x=df.drop('not.fully.paid',axis=1)
y=df['not.fully.paid']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
y_train.value_counts()
```
