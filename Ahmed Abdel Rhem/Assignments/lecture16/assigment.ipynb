```python
df=pd.read_csv('datasets/visitor-interests.csv')
df.head()
```

```python
import user_agents
```

```python
def extract_browser(x):
    ua = user_agents.parse(x)
    return pd.Series((ua.browser.family, ua.os.family, ua.device.family))
df[['Browser','OS','Device']]  = df['UserAgent'][0:51:1].apply(extract_browser)
df.head(60)
```

```python
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo
```

```python
df['Latitude']=df['IP'][0:51:1].apply(lambda x:ip2geo.get(x,api_key='free').latitude)
```

```python
df.head()
```

```python
df['Longitude']=df['IP'][0:51:1].apply(lambda x:ip2geo.get(x,api_key='free').longitude)
df.head()
```

```python
from geopy.distance import great_circle
def distance(x):
    return great_circle((x['Latitude'],x['Longitude']),(26.8206,30.8025)).kilometers
df['Dist_From_Egypt'] = df[2:51:1].apply(distance,axis=1)
```
