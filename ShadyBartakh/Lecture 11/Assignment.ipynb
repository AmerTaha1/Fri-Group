```python
# For first time use (write)
with open('weather.csv','w',newline='') as csvfile:
    writer = csv.DictWriter(csvfile,fieldnames=['Country','City','Local Time','weather','humidity','visibility','pressure','wind speed'])
    writer.writeheader()
    mydict={'Country':wthr_data['location']['country'],
            'City':wthr_data['location']['name'],
            'Local Time':wthr_data['location']['localtime'],
            'weather':wthr_data['current']['temperature'],
            'humidity':wthr_data['current']['humidity'],
            'visibility':wthr_data['current']['visibility'],
            'pressure':wthr_data['current']['pressure'],
            'wind speed':wthr_data['current']['wind_speed']}
    writer.writerow(mydict)
```

```python
# For the loops use (append)
with open('weather.csv','a',newline='') as csvfile:
    writer = csv.DictWriter(csvfile,fieldnames=['Country','City','Local Time','weather','humidity','visibility','pressure','wind speed'])
    mydict={'Country':wthr_data['location']['country'],
            'City':wthr_data['location']['name'],
            'Local Time':wthr_data['location']['localtime'],
            'weather':wthr_data['current']['temperature'],
            'humidity':wthr_data['current']['humidity'],
            'visibility':wthr_data['current']['visibility'],
            'pressure':wthr_data['current']['pressure'],
            'wind speed':wthr_data['current']['wind_speed']}
    writer.writerow(mydict)
```
