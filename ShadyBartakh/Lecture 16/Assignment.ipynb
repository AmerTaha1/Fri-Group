```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading Data

```python
df=pd.read_csv('50ulke.csv')
df.head(7)
```

### EDA

```python
df.info()
```

```python
df.describe()
```

```python
# Unique Continents
df['Continent'].value_counts()
```

```python
# Highest Income for a country
df[df['National Income']==df['National Income'].max()].Countries
```

```python
# Cleaning an unnecessary column of indices
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# Show most 5 countries data with highest wages.
df.sort_values(by='Per person', ascending=False).head()
```

### Analysis & Viz

```python
%matplotlib inline
```

```python
sns.set(rc={'figure.figsize': [15, 7]}, font_scale=1.3)
ax=sns.catplot(x='Countries',y='Growth',data=df,kind='bar')
plt.xticks(rotation=90)
```

```python
# Correlation between wages and the national income
sns.set(rc={'figure.figsize': [23, 15]}, font_scale=1.3)
sns.pairplot(df,vars=['National Income','Per person'],hue='Continent')
```

```python
# Minimum wages in Countries
sns.set(rc={'figure.figsize': [23, 15]}, font_scale=1.5)
sns.barplot(data=df,x='Countries',y='Per person',estimator=np.min)
plt.xticks(rotation=90)
```
