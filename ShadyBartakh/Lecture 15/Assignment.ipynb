```python
df=pd.read_csv('covid-19.csv')
df.sample(10)
```

```python
# Confirmed levels in regions
df.groupby('WHO Region')['Confirmed'].mean().sort_values(ascending=False)
```

```python
pd.pivot_table(df[(df['WHO Region']=='South-East Asia')], values=['Active'], index=['WHO Region'],
               columns=['Country/Region'], aggfunc=np.sum)
```
