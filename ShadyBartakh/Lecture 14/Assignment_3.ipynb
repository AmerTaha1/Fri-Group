```python
# 14)
cur.execute('SELECT books.Book_name, books.category_name, books.Rate, books.Price\
            FROM books FULL OUTER JOIN categories ON books.category_name=categories.category_name').fetchall()
```

```python
# 14) using the UNION ALL and LEFT JOIN clauses to emulate the SQLite FULL OUTER JOIN clause
cur.execute('SELECT b.Book_name, b.category_name, b.Rate, b.Price\
            FROM books b LEFT JOIN categories c ON b.category_name=c.category_name\
            UNION ALL SELECT b.Book_name, b.category_name, b.Rate, b.Price\
            FROM categories c LEFT JOIN books b ON b.category_name=c.category_name\
            WHERE b.category_name IS NULL;').fetchall()

"""
The UNION ALL clause retains the duplicate rows from the result sets of both queries.
The WHERE clause in the second SELECT statement;
removes rows that already included in the result set of the first SELECT statement.
"""
```
