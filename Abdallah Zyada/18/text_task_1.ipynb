```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
rest_df = pd.read_csv('dastasets/Restaurant_Reviews.csv', sep='\t')
```

```python
rest_df.head()
```

```python
victorizer = CountVectorizer(stop_words='english')
vector = victorizer.fit_transform(rest_df['Review'])
vector.toarray()
```

```python
vect_rst_df = pd.DataFrame(vector.toarray(), columns=victorizer.get_feature_names())
```

```python
vect_rst_df
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
vector = vectorizer.fit_transform(rest_df['Review'])
```

```python
tfidf_rst_df = pd.DataFrame(vector.toarray(), columns=vectorizer.get_feature_names())
```

```python
tfidf_rst_df
```
