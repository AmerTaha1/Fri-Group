```python
df=pd.read_csv('50_Startups.csv')
df.head()
```

```python
df.info()
```

```python
df[df['R&D Spend']==0]
```

```python
df.loc[df['R&D Spend']==0,'R&D Spend'] = df['R&D Spend'].median()
df.describe()
```

```python
df.loc[df['Marketing Spend']==0,'Marketing Spend']
df.loc[df['Marketing Spend']==0,'Marketing Spend'] = df['Marketing Spend'].median()
```

```python
df['Profit'].value_counts().sum()
```

```python
df.describe()
```

```python
sns.pairplot(df)
```

```python
df.columns
```

```python
dict = {'New York' : 1, 'California' : 2, 'Florida' : 3}
df['State_num'] = df['State'].map(dict)
df.head(10)
```

```python
x=df[['R&D Spend', 'Marketing Spend', 'State_num'  ]]
x
```

```python
y=df['Profit']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from xgboost import XGBRFRegressor
```

```python
models={
    'LR':LinearRegression(),
    'KNN':KNeighborsRegressor(),
    'SVR':SVR(),
    'DT':DecisionTreeRegressor(),
    'XGBR':XGBRFRegressor(),
    'RF':RandomForestRegressor()}
```

```python
for name,model in models.items():
    print(f'using:{name}')
    model.fit(x_train,y_train)
    print(f'Train Accuracy:{model.score(x_train,y_train)}')
    print(f'Test Accuracy:{model.score(x_test,y_test)}')
    print('-'*30,'\n')
```

```python
model=XGBRFRegressor()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
data=scaler.transform([[285349.20,500000,1]])
data
```

```python
prd=model.predict(data)
prd
```

```python
import joblib
joblib.dump(model,'model_Q1.h5')
joblib.dump(scaler,'scaler_Q1.h5')
```
