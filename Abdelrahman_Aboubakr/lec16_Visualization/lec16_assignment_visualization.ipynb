```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
```

```python
df = pd.read_csv('50ulke.csv')
df.head(5)
```

# Analysis on 50ulke and some Viz

## Highest National Income

```python
df[df['National Income'] == df ['National Income'].max()]['Countries']
```

## Lowest Per Person

```python
df[df['Per person'] == df ['Per person'].min()]['Countries']
```

## Plot Showing Countries VS National Income

```python
sns.set(rc={'figure.figsize':[16,8]}, font_scale=1.3)
sns.barplot(x = 'Countries', y = 'National Income', data = df)
plot.xticks(rotation=90)
```

## Viz2

```python
sns.boxplot(x = 'Continent', y = 'Growth', data = df)
plot.xticks(rotation=90)
```

## Viz3

```python
sns.swarmplot(x = 'Continent', y = 'National Income', data = df)
```
