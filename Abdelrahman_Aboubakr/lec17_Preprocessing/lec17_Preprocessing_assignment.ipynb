# Task2

```python
df = pd.read_csv('dastasets/ufo.csv')
df.head()
```

```python
df['Time']=pd.to_datetime(df['Time'],format='%m/%d/%Y %H:%M')
df.head()
```

```python
df['month']=df['Time'].dt.month
df.head()
```

```python
def in_dec(x):
    if x == 12:
        return 1
    else:
        return 0
df['In_December'] = df['month'].apply(in_dec)
df.head()
```

```python
from datetime import datetime
```

```python
df['Time_Diff_From_WW'] = (df['Time']-datetime(1914,7,28)) / np.timedelta64(1,'Y')
df.head()
```

```python
df.drop('month',inplace=True,axis=1)
```
