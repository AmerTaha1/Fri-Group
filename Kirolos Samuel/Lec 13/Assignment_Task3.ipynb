## Query 14 : Write a sql query to join both books & categories table into one new table containing book_name & category_name & book_rate and book_price

```python
cursor.execute("SELECT B.Book_Name,Cat.Category_name,B.Book_Rate,B.Book_Price FROM Books AS B , Category AS Cat WHERE B.Category_ID = Cat.Category_id ")
connection.commit()
rows = cursor.fetchall()
for row in rows:
    print(row)
```

## Query 15 : Write a sql query to calculate how many books each rate has and have price between 20 and 30

```python
cursor.execute("SELECT Count(*) FROM Books WHERE Book_Price BETWEEN 20 and 30 GROUP BY Book_Rate")
connection.commit()
rows = cursor.fetchall()
for index,row in enumerate(rows):
    print("The number of books that has the {} stars rating is {}.".format(index+1,row[0]))
```

## Query 16 : Write a sql query to calculate how many books each category has having count > 10.

```python
cursor.execute("SELECT Cat.Category_name,Count(*) FROM Books AS B,Category AS Cat WHERE B.Category_ID = Cat.Category_id GROUP BY B.Category_ID HAVING COUNT(*)>10")
connection.commit()
rows = cursor.fetchall()
for category_name,book_no in rows:
    print("The {} category has {} books in the library database.".format(category_name,book_no))
```
