```python
import requests
from bs4 import BeautifulSoup
```

```python
book_dict_rat={'Zero':0,'One':1,'Two':2,'Three':3,'Four':4,'Five':5}
book_dict_rat
```

```python
url='https://books.toscrape.com/index.html'
response=requests.get(url)
response=BeautifulSoup(response.text,'html.parser')
response
```

## First Way , easy to scrap book names and prices but impossible to scrap the categories

```python
book_names=[]
for res in response.find_all('h3'):
    book_name=res.find('a').get('title')
    book_names.append(book_name)
book_names
```

```python
book_prices=[]
for res in response.find_all('div',attrs={'class':'product_price'}):
    book_price=res.find('p').get_text()
    book_prices.append(book_price)
book_prices
```

## Second and the most efficient way

```python
main_url="".join(url.split('/')[:-1])+"/"
main_url=main_url[0:6]+'//'+main_url[6:]
main_url
```

```python
book_urls=[]
for res in response.find_all('h3'):
    book_url=res.find('a').get('href')
    book_urls.append(main_url+book_url)
book_urls
```

```python
book_names=[]
book_prices=[]
book_categories=[]
book_ratings=[]
for url in book_urls:
    resp=requests.get(url)
    resp=BeautifulSoup(resp.text,'html.parser')
    #book category
    book_categ=resp.find_all('a')[3].get_text()
    #book name
    book_name=resp.find('img').get('alt')
    #book price
    book_price=resp.find('div',attrs={'class':'col-sm-6 product_main'}).find('p').get_text()[1:]
    #book rate
    book_rate=book_dict_rat[resp.find('div',attrs={'class':'col-sm-6 product_main'}).find_all('p')[2].get('class')[1]]
    book_ratings.append(book_rate)
    book_names.append(book_name)
    book_prices.append(book_price)
    book_categories.append(book_categ)
```

```python
import csv
```

```python
with open('books.csv','w',encoding='utf-8',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(['Book_name','Price','Category','Rate','Link'])
    for Book_name,Price,Category_name,Rate,URL in zip(book_names,book_prices,book_categories,book_ratings,book_urls):
        writer.writerow([Book_name,Price,Category_name,Rate,URL])
```
