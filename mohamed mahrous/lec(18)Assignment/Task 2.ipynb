```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="most_frequent")

df["Product_Category_2"] = imputer.fit_transform(df[["Product_Category_2"]])
df["Product_Category_3"] = imputer.fit_transform(df[["Product_Category_3"]])
df.isnull().sum()
```

```python
df.shape
```

```python
df.info()
```

```python
df["Product_ID"] = df["Product_ID"].apply(lambda x : int(x[1:]))
df.head()
```

```python
df["City_Category"].value_counts()
```

```python
df["Stay_In_Current_City_Years"].value_counts()
```

```python
df["Marital_Status"].value_counts()
```

```python
df["Marital_Status"] = pd.to_numeric(df["Marital_Status"], errors="coerce")
```

```python
df = pd.get_dummies(df, columns=["Age", "City_Category", "Stay_In_Current_City_Years", "Gender"], drop_first=True)
df.head()
```

# loan_data.csv Dataset

```python
df = pd.read_csv("../session/dastasets/loan_data.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df = pd.get_dummies(df, columns=["purpose"], drop_first=True)
df.head()
```

```python
df.describe()
```

```python
from datasist.structdata import detect_outliers

for col in df.columns:
    outliers_indices = detect_outliers(df, 0, [col])
    df[col].loc[outliers_indices] = df[col].median()
```
