```python
import pandas as pd
```

# CountVectorizer

```python
df = pd.read_csv("../session/dastasets/Restaurant_Reviews.csv", sep="\t")
df.head()
```

```python
df.shape
```

```python
df["Review"][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(stop_words="english")
vector = vectorizer.fit_transform(df["Review"])

df2 = pd.DataFrame(data=vector.toarray(), columns=vectorizer.get_feature_names())
df2.head()
```

# TfidfVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words="english")
vector = vectorizer.fit_transform(df["Review"])

df3 = pd.DataFrame(data=vector.toarray(), columns=vectorizer.get_feature_names())
df3.head()
```
