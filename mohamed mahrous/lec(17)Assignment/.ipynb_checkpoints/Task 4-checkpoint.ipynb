```python
import pandas as pd
import user_agents
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo
```

# Creating ["Browser", "OS", "Device"] columns

```python
df = pd.read_csv("../session/dastasets/visitor-interests.csv")
df.head()
```

```python
df.shape
```

```python
_50raw_df = df.iloc[:50]
_50raw_df
```

```python
def Extract_UserAgent_Info(x):
    ua = user_agents.parse(x)
    return pd.Series((ua.browser.family, ua.os.family, ua.device.family))

_50raw_df[["Browser","OS","Device"]] = _50raw_df["UserAgent"].apply(Extract_UserAgent_Info)
_50raw_df.head()
```

# Creating ["Latitude", "Longitude"] columns

```python
def extract_lat_long(x):
    response = ip2geo.get(x, api_key="free")
    return pd.Series((response.latitude, response.longitude))

_50raw_df[["Latitude", "Longitude"]] = _50raw_df["IP"].apply(extract_lat_long)
_50raw_df.head()
```
