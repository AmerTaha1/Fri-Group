```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../session/dastasets/titanic_full.csv")
df.head()
```

```python
df.shape
```

# Creating "Title" column

```python
df["Title"] = df["Name"].apply(lambda x : x.split(",")[1].split(".")[0][1:])
df.head()
```

```python
df["Title"].value_counts()
```

```python
sns.set(rc={"figure.figsize":[20,5]}, font_scale = 1.3)
sns.countplot(x = "Title", data=df)
```
