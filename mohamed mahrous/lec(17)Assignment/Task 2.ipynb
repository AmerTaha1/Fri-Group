```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
df = pd.read_csv("../session/dastasets/ufo.csv")
df.head()
```

```python
df["Time"] = pd.to_datetime(df["Time"], format="%m/%d/%Y %H:%M")
df.head()
```

```python
df.info()
```

# Creating "IN_December" column

```python
df["IN_December"] = df["Time"].dt.month.apply(lambda x : 1 if x == 12 else 0)
df
```

# Creating "Time_Diff_From_WW1" column

```python
df["Time_Diff_From_WW1"] = np.abs((datetime(1914,7,28) - df["Time"])/np.timedelta64(1,"Y"))
df.head()
```
