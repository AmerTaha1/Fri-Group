### data visualization with many aspects


```python
sns.distplot(df['Growth'])
```

```python
sns.kdeplot(df['Per person'],shade=True)
```

```python
sns.jointplot(x='Growth',y='National Income',data=df)
```

```python
sns.pairplot(df)
```

```python
sns.boxplot(x='Continent',y='Per person',data=df)
```

```python
sns.violinplot(x='Continent',y='Per person',data=df)
```

```python
sns.stripplot(x='Continent',y='Per person',data=df)
```

```python
sns.swarmplot(x='Continent',y='Growth',data=df)
```

```python
sns.barplot(x='Continent',y='Growth',data=df , estimator=np.max)
```

```python
sns.countplot(x='Continent',data=df)
sns.set(rc={'figure.figsize':[8, 8]},font_scale=1.3)
```

```python
sns.countplot(df['Growth'])
```

```python
sns.catplot(x='Countries',col='Continent',kind='count',data=df)
```
