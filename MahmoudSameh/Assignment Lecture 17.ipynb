```python
import numpy as np
from geopy.distance import great_circle
def extract_dist(x):
    try:
        return great_circle((x['latitude'],x['longitude']),(30.033333,31.233334)).kilometers
    except ValueError:
        return np.nan
df['Dist_From_Egypt']=df.apply(extract_dist,axis=1)
df.head(50)
```
