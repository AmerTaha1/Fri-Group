# Experiment 2: LUT dataset Analysis.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()

adt_lut = pd.read_csv('LUT.csv')
adt_lut.head()
```

## keep L & VDS & VSB constant to see how ID change with VGS

```python
adt_lut.L.value_counts(), adt_lut.VDS.value_counts(), adt_lut.VSB.value_counts()
```

```python
L_cons, VDS_cons, VSB_cons =  2.000000e-06, 0.900, 0.05
adt_lut_v1 = adt_lut[(adt_lut.L == L_cons) & (adt_lut.VDS == VDS_cons) & (adt_lut.VSB == VSB_cons)]
sns.jointplot(x="VGS", y="ID", data=adt_lut_v1)
plt.show()
```
