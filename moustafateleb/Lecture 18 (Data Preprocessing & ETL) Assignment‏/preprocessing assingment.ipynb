## Preprocesing Data using modules

```python
df = pd.read_csv('dastasets/loan_data.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df=pd.get_dummies(df,columns=['purpose'],drop_first=True)
df.head()
```

```python
cols=['installment','int.rate','log.annual.inc']

sns.boxplot(x='log.annual.inc',data=df)
sns.stripplot(x='log.annual.inc',data=df)
```

```python

from datasist.structdata import detect_outliers

for col in cols:
    indices=detect_outliers(df,0,[col])
    df.loc[indices,col]=df[col].median()
```

```python
df['not.fully.paid'].value_counts()
```

```python
x=df.drop('not.fully.paid',axis=1)
y=df['not.fully.paid']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
y_train.value_counts()
```
