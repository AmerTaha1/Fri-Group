```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('black_friday.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().any()
```

```python
df.dropna(subset=['Product_Category_2'],inplace=True)
df.dropna(subset=['Product_Category_3'],inplace=True)
```

```python
df.isnull().any()
```

```python
df['Product_ID']=df['Product_ID'].apply(lambda x : x.split('P')[1])
df
```

```python
df['Product_Category_2']=pd.to_numeric(df['Product_Category_2'],errors='coerce')
df['Product_Category_3']=pd.to_numeric(df['Product_Category_3'],errors='coerce')
df['Product_ID']=pd.to_numeric(df['Product_ID'],errors='coerce')
df.info()
```

```python
sns.boxplot(x='Purchase', data=df)
sns.stripplot(x='Purchase', data=df)
```
