## Date&Time

```python
df=pd.read_csv('datasets/ufo.csv')
df.head()
```

```python
type(df['Time'][0])
```

```python
df.info()
```

```python
df['Time']=pd.to_datetime(df['Time'],format='%m/%d/%Y %H:%M')
df.head()
```

```python
df.info()
```

```python
df['month']=df['Time'].dt.month
df.head()
```

```python
df['month_name']=df['Time'].dt.month_name()
df.head()
```

```python
df['day']=df['Time'].dt.day
df['Day_name']=df['Time'].dt.day_name()
df.head()
```

```python
df['year']=df['Time'].dt.year
df.head()
```

```python
df['hour']=df['Time'].dt.hour
df['minute']=df['Time'].dt.minute
df.head()
```

```python
def map_hour(x):
    if x in range(0,13):
        return 'Morning'
    elif x in range(13,19):
        return 'Afternoon'
    else:
        return 'Evening'
map_hour(14)
```

```python
df['Period']=df['hour'].apply(map_hour)
df.head()
```

```python
sns.countplot(x='Period',data=df)
```

```python
sns.countplot(x='Day_name',data=df)
```

```python
sns.countplot(x='Day_name',data=df,hue='Period')
```

```python
sns.countplot(x='month_name',data=df)
```

```python
sns.set(rc={'figure.figsize': [12, 12]}, font_scale=1.3)
```

```python
sns.countplot(y='month_name',data=df,hue='Day_name')
```

```python
def map_month(x):
    if x in [12,1,2]:
        return 'Winter'
    elif x in [3,4,5]:
        return 'Spring'
    elif x in [6,7,8]:
        return 'Summer'
    else:
        return 'Autumn'
df['season']=df['month'].apply(map_month)
df.head()
```

```python
sns.countplot(x='season',data=df)
```

```python
sns.countplot(hue='season',y='year',data=df)
```

```python
df['Elapsed_years']=(datetime(1973,10,6)-df['Time'])/np.timedelta64(1,'Y')#M,D,W,h,m
df.head()
```

```python
def extract_cctype(x):
    x=str(x)
    if x.startswith('1'):
        return 'Travel & Business'
    elif x.startswith('2'):
        return 'Visa'
    elif x.startswith('3'):
        return 'Master Card'
    elif x.startswith('4'):
        return 'Paypal'
    elif x.startswith('5'):
        return 'Discover'
    else:
        return 'mmmmm'

df['cc_type']=df['Credit Card'].apply(extract_cctype)
df.head()
        
```

```python
sns.countplot(x='cc_type',data=df)
```

```python
def extract_browser(x):
    
    ua=user_agents.parse(x)
    return pd.Series((ua.browser.family,ua.os.family,ua.device.family))

df[['Browser','Os','Device']]=df['Browser Info'].apply(extract_browser)
df.head()
```

## Geospatial data

```python
df=pd.read_csv('datasets/sendy_logistics.csv').sample(200)
df.head()
```

```python
from geopy.distance import great_circle
df.info()
```

```python
df['Pickup Lat']
```

```python
df['Pickup Long']
```

```python
df['Destination Lat']
```

```python
df['Destination Long']
```

```python
great_circle((-1.317755,36.830370),(-1.300406,36.829741)).kilometers
```

```python
def extract_dist(x):
    return great_circle((x['Pickup Lat'],x['Pickup Long']),(x['Destination Lat'],x['Destination Long'])).kilometers
df['Distance (km)']=df.apply(extract_dist,axis=1)
df.head()
```

```python
from geopy.geocoders import Nominatim
u=Nominatim(user_agent='mm')
loc=u.geocode('Cairo Festival , Cairo , Egypt')
loc.raw
```

```python
from geopy.geocoders import Nominatim
geolocator=Nominatim(user_agent='cccc')
```

```python
loc=u.reverse('-1.317755,36.830370')
loc.raw['address']['country']
```

```python
def ff(x):
    
    return geolocator.reverse(f"{x['Destination Lat']},{x['Destination Long']}").raw['address']['country']

df['Destination_country']=df.apply(ff,axis=1)
df.head()
```
