```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import user_agents
```

```python
df = pd.read_csv('visitor-interests.csv')
df.head()
```

```python
def Browser(x):
    for x in df['UserAgent'][0:49]:
        return x.split('/')[0]
```

```python
df['Browser']=df['UserAgent'].apply(Browser)
df.head()
```

```python
def OS(x):
    for x in df['UserAgent'][0:49]:
        user_agents.parse(x)
        return x.split('/')[1]
```

```python
df['OS']=df['UserAgent'].apply(OS)
df.head()
```

```python
for x in df['UserAgent'][:49]:
    ua=user_agents.parse(x)
    ua.browser.family
```

```python
def UserAgents(x):
    
    ua=user_agents.parse(x)
    return pd.Series((ua.browser.family,ua.os.family,ua.device.family))

df[['Browser','Os','Device']]=df['UserAgent'].apply(UserAgents)
df.head()
```
