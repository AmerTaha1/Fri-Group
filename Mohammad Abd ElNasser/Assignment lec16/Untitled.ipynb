### Discrebtion on 50ulke dataset
This dataset talking about the economy of countries.
We have 5 columns 
1- unnammed
2- Countries
3- National Income	
4- Growth
5- Continent
we gonna make some relations between them 

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('50ulke.csv')
df.head()
```

# Viz 1
showin some info about the growth

```python
sns.distplot(df['Growth'], kde=False) 
```

# Viz 2 
showwing some relations between ( National Income ) & ( Per person )

```python
sns.pairplot(df, vars=['National Income', 'Per person'])
```

# Viz 3
Here , we are showing how much National Income is every country making 

```python
sns.barplot(x='Countries', y='National Income', data=df)
```

```python
sns.barplot(x='Countries', y='National Income', data=df, estimator=np.max)
```

```python
sns.barplot(x='Countries', y='National Income', data=df, estimator=np.min)
```

# Viz 4 
This vizualization like the Viz 3 but we here see the relation between the viz 3 and Continent

```python
sns.barplot(x='Countries', y='National Income', data=df, hue='Continent')
```

# Viz 5 
here we show the relation between the national income in each country and the how it is for each person 

```python
df=df.pivot_table(index='Countries', columns='National Income', values='Per person')
df
```

```python
sns.heatmap(df)
```
