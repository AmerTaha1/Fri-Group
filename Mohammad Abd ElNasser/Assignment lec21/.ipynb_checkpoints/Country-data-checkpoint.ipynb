# Here, We can't use Scaler cuz we have catogrical data ( country ) so we will drop it and try again

```python
df.drop('country', axis=1, inplace =True)
df
```

```python
scaler = StandardScaler()
x=scaler.fit_transform(df)
x
```

# now it's okay so we gonna try our methods in clustring

```python
# Elbow Method
wcss = []
for i in range(1,150):
    model = KMeans(n_clusters=i)
    clusters=model.fit_predict(x)
    wcss.append(model.inertia_)

plt.plot(range(1,150),wcss)
plt.xlabel('n_clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
```

```python
model=KMeans(n_clusters=80)
clusters=model.fit_predict(x)
df['Clusters']=clusters
df
```

```python
centroids=model.cluster_centers_
centroids
```

```python
plt.scatter(x[:,1],x[:,8],c=clusters,cmap='viridis')
plt.scatter(centroids[:,1],centroids[:,8], s=100,c='red')
```

# Let's try the other one 

```python
df = pd.read_csv("Country-data.csv")
df
```

```python
df.drop('country', axis=1, inplace =True)
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x=scaler.fit_transform(df)
x
```

```python
import scipy.cluster.hierarchy as sch
dendrogram =sch.dendrogram(sch.linkage(x,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=2)
clusters=model.fit_predict(x)
df['Clusters'] = clusters
df
```

```python
plt.scatter(x[:,3],x[:,8],c=clusters,cmap='viridis')
```

```python
x = df.drop('Clusters', axis=1)
y =df['Clusters']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
models={
    'LR':LogisticRegression(),
    'KNN':KNeighborsClassifier(),
    'DT':DecisionTreeClassifier(),
    'SVC':SVC(),
    'NB':GaussianNB(),
    'XGC':XGBClassifier(),
    'RF':RandomForestClassifier()
    
}
```

```python
for name,model in  models.items():
    print(f'using {name}: ')
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(f'Training Accuracy :{accuracy_score(y_train,model.predict(x_train))}')
    print(f'Testing Accuracy :{accuracy_score(y_test,y_pred)}')
    print(f'Confusion matrix:\n {confusion_matrix(y_test,y_pred)}')
    print(f'Recall: {recall_score(y_test,y_pred)}')
    print(f'precision: {precision_score(y_test,y_pred)}')
    print(f'F1-score: {f1_score(y_test,y_pred)}')
    print(f'Fbeta-score: {fbeta_score(y_test,y_pred,beta=0.5)}')
    print(classification_report(y_test,y_pred))
    print('-'*33)
    
```
